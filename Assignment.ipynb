# Classifications Vs Petal & Sepel Measurement

```python
plt.figure(figsize=(20,10))
plt.xlabel('Measurements',fontsize=15)
plt.ylabel(' ')
plt.title('Classification Vs Petal & Sepal Measurements',fontsize=25)
plt.scatter(df['SL'],df['Classification'],c='b')
plt.scatter(df['PW'],df['Classification'],c='y')
plt.scatter(df['SW'],df['Classification'],c='g')
plt.scatter(df['PL'],df['Classification'],c='k')
plt.legend(df)
plt.show()
```

# Classifications Vs Sepal Width

```python
sns.set_palette('bright')
plt.xlabel(' Sepal Width')
plt.ylabel(' ')
plt.title('Classification Vs Sepal Width')
sns.scatterplot(df['SW'],df['Classification'])

plt.show()
```

# Classifications Vs Sepal Length

```python
sns.set_palette('bright')
ax=sns.boxplot(x=df['Classification'],y=df['SL']);
ax.set_xlabel(" ")
ax.set_ylabel("Sepal Length")
plt.title('Classification Vs Sepal Length',fontsize=15)
plt.show()
```

# Classifications Vs Petal Length

```python
sns.set_palette('pastel6')
ay=sns.barplot(x=df.Classification,y=df.PL);
ay.set_xlabel(" ")
ay.set_ylabel("Petal Length")
plt.title('Classification Vs Petal Length',fontsize=15)
plt.show()
```

# Classifications Vs Petal Width

```python
sns.set_palette('bright')
aw=sns.barplot(y=df['Classification'],x=df['PW'],orient='h')
aw.set_xlabel("Petal Width")
aw.set_ylabel("")
plt.title('Classification Vs Petal Width',fontsize=15)
plt.show()
```
